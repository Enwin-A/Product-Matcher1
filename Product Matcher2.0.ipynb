```python
#have run this for 13 different products and it would take 16 minutes to scrape 13 different products
#this is the code for the searcher and scraper of urls and descriptions
from selenium import webdriver
import json
import time
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys

def search_ulta(product_name):
    # Set up the Chrome driver
    url = 'https://www.ulta.com/'
    chrome_options = Options()
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument("--start-maximized")
    chrome_options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(executable_path='chromedriver.exe', options=chrome_options)
    driver.get(url)

    # Find the search bar and input the product name
    search_bar = driver.find_element_by_class_name("InputFieldUnderline__input")
    search_bar.send_keys(product_name)
    search_bar.send_keys(Keys.RETURN)

    # Allow time for the search results to load
    time.sleep(5)
    product_urls = []

    # Load existing data from the JSON file
    existing_data = []
    try:
        with open('ulta.json', 'r', encoding='utf-8') as existing_file:
            existing_data = json.load(existing_file)
    except FileNotFoundError:
        pass

    # Create a set of unique product names from the existing data
    unique_product_names = set(item.get('ProductName', 'N/A') for item in existing_data)

    # Create a list to store new data
    new_data_list = []

    # Extract and print the URLs of the search results
    search_results = driver.find_elements_by_class_name("ProductCard")
    for result in search_results:
        header = result.find_element_by_class_name("Link_Huge")
        if header:
            print("true")
            product_url = header.get_attribute("href")
            print(product_url)
            product_urls.append(product_url)

    for i in product_urls:
        driver.get(i)
        time.sleep(2)  # Allow time for the page to load

        # Extract information from the product page
        box = driver.find_element_by_class_name("ProductInformation")
        if box:
            title = box.find_element_by_xpath('//*[@id="92384e5c-2234-4e8f-bef7-e80391889cfc"]/h1')
            if title:
                product_title = title.find_element_by_class_name("Link_Huge").text
                product_name1 = title.find_element_by_xpath('//*[@id="92384e5c-2234-4e8f-bef7-e80391889cfc"]/h1/span[2]').text

        box2 = driver.find_element_by_class_name("ProductSummary")
        if box2:
            product_description = box2.find_element_by_xpath('//*[@id="b46bc3ad-9907-43a6-9a95-88c160f02d7f"]/p').text

        # Check for duplicates before adding to the new data list
        if product_name1 not in unique_product_names:
            # Store the gathered information in a dictionary
            product_details = {
                'Search Result for': product_name,
                'ProductTitle': product_title,
                'ProductName': product_name1,
                'ProductDescription': product_description,
                'ProductURL': i  # product_url
            }

            # Append the dictionary to the new data list
            new_data_list.append(product_details)

            # Add the product name to the set of unique names
            unique_product_names.add(product_name1)

            # Print or store the gathered information
            print("product Title:", product_title)
            print("Product Name:", product_name1)
            print("Product Description:", product_description)
            print("-" * 50)
        time.sleep(2)

    # Combine existing data and new data
    updated_data = existing_data + new_data_list

    with open('ulta.json', 'w', encoding='utf-8') as output_file:
        json.dump(updated_data, output_file, ensure_ascii=False, indent=2)

    # Close the browser
    driver.quit()
    print("JSON File created")

# Example usage
import json

# Load the JSON file
json_file_path = 'Amazon.json'  # Use double backslashes or a raw string
with open(json_file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Iterate through each product in the JSON data
for product in data:
    product_name = product.get('ProductName', 'N/A')
    product_description = '\n'.join(product.get('productDescription', []))  # Combine multiple lines
    # Print or process the extracted information
    print(f"Product Name: {product_name}")
    search_ulta(product_name)
```

```python
#updated and working as of rn 4:40 pm 3rd december FINAL 
#this is more refined as in it only checks for the amazon datas results from ulta and based on the results we match the best one
#this is done only for a sample size and not much 
#this ones json file output is for only a few products and based on that the first one of each amazon product names match in ulta is the best match
#the output json file contains only the best matches


#the missing values have been handled and fixed and now all thats left to do is scale


import json
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
from fuzzywuzzy import fuzz

# Load Amazon and Ulta.com product data from JSON files
with open('Amazon.json', 'r', encoding='utf-8') as file:
    amazon_data = json.load(file)

with open('ulta.json', 'r', encoding='utf-8') as file:
    ulta_data = json.load(file)

# Create an empty list for the JSON output
output_data = []

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to get BERT embeddings for a given text
def get_bert_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy()
    return embeddings

# Matching algorithm based on fuzzy string matching and BERT embeddings
def match_products(amazon_product, ulta_product):
    # Check if 'ProductName' key exists in both amazon_product and ulta_product
    if 'ProductName' not in amazon_product or 'ProductName' not in ulta_product:
        return False

    amazon_name = amazon_product['ProductName']
    ulta_name = ulta_product['ProductName']
    print(amazon_name)
    print(ulta_name)
    
    
    if 'productDescription' in amazon_product:
        amazon_description = ' '.join(amazon_product['productDescription']) if isinstance(amazon_product['productDescription'], list) else amazon_product['productDescription']
#         amazon_description_words = set(amazon_description.lower().split())
    else:
        print("amazon desc = none")
#         amazon_description_words = set()

    if 'ProductDescription' in ulta_product:
        ulta_description = ulta_product['ProductDescription']
#         ulta_description_words = set(ulta_description.lower().split())
    else:
        print("ulta desc = none")
    
    # Get BERT embeddings for product names and descriptions
    amazon_name_embedding = get_bert_embeddings(amazon_name)
    ulta_name_embedding = get_bert_embeddings(ulta_name)
    
    amazon_description_embedding = get_bert_embeddings(amazon_description)
    ulta_description_embedding = get_bert_embeddings(ulta_description)

    # Calculate cosine similarity between BERT embeddings for names
    name_embedding_similarity_score = cosine_similarity([amazon_name_embedding], [ulta_name_embedding])[0][0]
    print("name embedding similarity score is:")
    print(name_embedding_similarity_score)

    # Calculate cosine similarity between BERT embeddings for descriptions
    description_embedding_similarity_score = cosine_similarity([amazon_description_embedding], [ulta_description_embedding])[0][0]
    print("description embedding similarity score is:")
    print(description_embedding_similarity_score)

    # Combine scores with a weighted sum
#     combined_score = 0.5 * name_similarity_score + 0.5 * description_embedding_similarity_score
    combined_score = 0.5 * name_embedding_similarity_score + 0.5 * description_embedding_similarity_score
    print("combined similarity score is:")
    print(combined_score)
    print("--------------------------------------NEXT---------------------------------------")
    # Set a threshold for combined score (adjust based on your needs)
    return combined_score #> 0.88  # Example threshold: 0.7

# Iterate through Amazon products and find matches in Ulta.com data
for amazon_product in amazon_data:
    # Check if 'ProductName' is present in the current Amazon product
    if 'ProductName' in amazon_product:
        amazon_name = amazon_product['ProductName']

        # Check if there is a corresponding Ulta product with a matching 'Search Result for'
        matching_ulta_products = [ulta_product for ulta_product in ulta_data if ulta_product.get('Search Result for') == amazon_name]

        # List to store scores of all matches
        scores = []

        # Iterate through matching Ulta products and calculate scores
        for ulta_product in matching_ulta_products:
            score = match_products(amazon_product, ulta_product)
            scores.append((ulta_product, score))

        # Check if there is at least one match
        if scores:
            # Get the Ulta product with the maximum score
            best_match = max(scores, key=lambda x: x[1])
            # Append only the match with the maximum score to the output
            output_data.append({
                'Amazon_ProductName': amazon_product.get('ProductName', 'N/A'),
                'Ulta_ProductName': best_match[0].get('ProductName', 'N/A'),
                'Ulta_URL': best_match[0].get('ProductURL', 'N/A'),
                'Score': best_match[1]
            })

# Save the output to a JSON file
print("JSON file created")
with open('matching_results.json', 'w', encoding='utf-8') as output_file:
    json.dump(output_data, output_file, ensure_ascii=False, indent=2)
```
